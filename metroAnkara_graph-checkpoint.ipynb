```python
import networkx as nx
import matplotlib.pyplot as plt

# Hatlardaki duraklar
A1 = ["AŞTİ", "Emek", "Bahçelievler", "Beşevler", "Maltepe", "Demirtepe", "Kızılay", "Kolej", "Kurtuluş", "Cebeci", "Dikimevi"]
M1 = ["Kızılay", "Sıhhiye", "Adliye", "Ulus", "Atatürk Kültür Merkezi", "Akköprü", "İvedik", "Yenimahalle", "Demetevler", "Hastane", "Macunköy", "OSTİM", "Batıkent"]
M2 = ["Kızılay", "Necatibey", "Milli Kütüphane", "Söğütözü", "MTA", "Emek", "Bilkent", "Beytepe", "Ümitköy", "Çayyolu", "Koru"]
M3 = ["Batıkent", "Batı Merkez", "Mesa", "Botanik", "İstanbul Yolu", "Eryaman 5", "Eryaman 1-2", "Devlet Mah.", "Harikalar Diyarı", "GOP", "OSB-Törekent"]
M4 = ["AKM", "ASKİ", "Dışkapı", "Meteoroloji", "Belediye", "Mecidiye", "Kuyubaşı", "Dutluk", "Şehitler"]
BAŞKENTRAY = ["Sincan", "Lale", "Elvankent", "Eryaman YHT", "Özgüneş", "Etimesgut", "Havadurağı", "Yıldırım", "Behiçbey", "Motor", "Gazi", "Gazi Mahallesi", "Hipodrom", "Ankara Gar", "Kurtuluş", "Cebeci", "Demirlibahçe", "Saimekadın", "Mamak", "Bağderesi", "Üreğil", "Köstence", "Kayas"]
T1 = ["Yenimahalle", "İvedik", "Yunus Emre", "TRT Seyir", "Şentepe"]

# Graf oluştur
G = nx.Graph()

# Hat bağlantılarını ekleyen fonksiyon
def add_line(line):
    for i in range(len(line) - 1):
        G.add_edge(line[i], line[i + 1])

# Hatları graf'a ekle
for line in [A1, M1, M2, M3, M4, BAŞKENTRAY, T1]:
    add_line(line)

# Görselleştirme
plt.figure(figsize=(18, 12))
nx.draw(G, with_labels=True, node_color="skyblue", node_size=800, font_size=8)
plt.title("Ankara Metro Sistemi - Graph Teorisi ile Modelleme", fontsize=14)
plt.show()
```

```python
import numpy as np

# Düğümleri liste halinde al
nodes = list(G.nodes())
nodes.sort()  # Alfabetik sıralama (isteğe bağlı)

# Düğüm isimlerini indeksle
node_index = {node: i for i, node in enumerate(nodes)}

# Komşuluk matrisi oluştur
adj_matrix = np.zeros((len(nodes), len(nodes)), dtype=int)

# Kenarları matrise işle
for u, v in G.edges():
    i, j = node_index[u], node_index[v]
    adj_matrix[i][j] = 1
    adj_matrix[j][i] = 1  # Simetrik çünkü yönsüz grafik

# Matrisi yazdır
print("Komşuluk Matrisi:")
print("    " + "  ".join([n[:4] for n in nodes]))  # İlk 4 harfi göster
for i, row in enumerate(adj_matrix):
    print(f"{nodes[i][:4]} {row}")
```

```python
import networkx as nx

# En kısa yol fonksiyonu
def en_kisa_yol(graf, kaynak, hedef):
    try:
        yol = nx.shortest_path(graf, source=kaynak, target=hedef)
        uzunluk = nx.shortest_path_length(graf, source=kaynak, target=hedef)
        return yol, uzunluk
    except nx.NetworkXNoPath:
        return None, float('inf')

# Örnek sorgu
kaynak_durak = "Kızılay"
hedef_durak = "Sincan"
yol, uzunluk = en_kisa_yol(G, kaynak_durak, hedef_durak)

# Sonuçları göster
print(f"{kaynak_durak} → {hedef_durak} arası en kısa yol ({uzunluk} durak):")
print(" → ".join(yol))
```

```python
def metro_yol_sorgulama(graf):
    print("\n📍 Ankara Metro En Kısa Yol Sorgulayıcısı 📍\n")
    print("Mevcut duraklar:")
    print(", ".join(sorted(graf.nodes())))
    print("\nBaşlangıç ve varış duraklarını giriniz.\n")

    kaynak = input("Başlangıç durağı: ").strip()
    hedef = input("Varış durağı: ").strip()

    if kaynak not in graf.nodes() or hedef not in graf.nodes():
        print("\n❌ Hatalı durak ismi girdiniz. Lütfen listeden seçiniz.")
        return

    yol, uzunluk = en_kisa_yol(graf, kaynak, hedef)
    if yol:
        print(f"\n✅ {kaynak} → {hedef} arası en kısa yol ({uzunluk} durak):")
        print(" → ".join(yol))
    else:
        print(f"\n🚫 {kaynak} → {hedef} arasında yol bulunamadı.")

# Terminal çalıştırma
metro_yol_sorgulama(G)
```

```python
import tkinter as tk
from tkinter import ttk
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

# Metro hatları
hatlar = {
    "A1": ["AŞTİ", "Emek", "Bahçelievler", "Beşevler", "Maltepe", "Demirtepe", "Kızılay", "Kolej", "Kurtuluş", "Cebeci", "Dikimevi"],
    "M1": ["Kızılay", "Sıhhiye", "Adliye", "Ulus", "Atatürk Kültür Merkezi", "Akköprü", "İvedik", "Yenimahalle", "Demetevler", "Hastane", "Macunköy", "OSTİM", "Batıkent"],
    "M2": ["Kızılay", "Necatibey", "Milli Kütüphane", "Söğütözü", "MTA", "Emek", "Bilkent", "Beytepe", "Ümitköy", "Çayyolu", "Koru"],
    "M3": ["Batıkent", "Batı Merkez", "Mesa", "Botanik", "İstanbul Yolu", "Eryaman 5", "Eryaman 1-2", "Devlet Mah.", "Harikalar Diyarı", "GOP", "OSB-Törekent"],
    "M4": ["Atatürk Kültür Merkezi", "ASKİ", "Dışkapı", "Meteoroloji", "Belediye", "Mecidiye", "Kuyubaşı", "Dutluk", "Şehitler"],
    "BAŞKENTRAY": ["Sincan", "Lale", "Elvankent", "Eryaman YHT", "Özgüneş", "Etimesgut", "Havadurağı", "Yıldırım", "Behiçbey", "Motor", "Gazi", "Gazi Mahallesi", "Hipodrom", "Ankara Gar", "Kurtuluş", "Cebeci", "Demirlibahçe", "Saimekadın", "Mamak", "Bağderesi", "Üreğil", "Köstence", "Kayas"],
    "T1": ["Yenimahalle", "İvedik", "Yunus Emre", "TRT Seyir", "Şentepe"]
}

# Graf yapısı oluştur
G = nx.Graph()
for hat in hatlar.values():
    for i in range(len(hat) - 1):
        G.add_edge(hat[i], hat[i + 1])

# Sabit konumlar sütun gibi
def sabit_konumlar_sutun(hatlar_dict):
    pos = {}
    x_step = 5
    y_step = -1
    x = 0
    for hat in hatlar_dict.values():
        for i, durak in enumerate(hat):
            if durak not in pos:
                pos[durak] = (x, i * y_step)
        x += x_step
    return pos

# En kısa yolu bul
def en_kisa_yol(graf, kaynak, hedef):
    try:
        yol = nx.shortest_path(graf, source=kaynak, target=hedef)
        return yol, len(yol) - 1
    except nx.NetworkXNoPath:
        return None, None

# Arayüz
root = tk.Tk()
root.title("Ankara Metro - Rota ve Harita")
root.state('zoomed')  # Tam ekran

frame = ttk.Frame(root)
frame.pack(pady=20)

ttk.Label(frame, text="Başlangıç Durağı:").grid(row=0, column=0, padx=5, pady=5)
ttk.Label(frame, text="Varış Durağı:").grid(row=1, column=0, padx=5, pady=5)

duraklar = sorted(G.nodes())
start_var = tk.StringVar()
end_var = tk.StringVar()
start_combo = ttk.Combobox(frame, textvariable=start_var, values=duraklar, width=30)
end_combo = ttk.Combobox(frame, textvariable=end_var, values=duraklar, width=30)
start_combo.grid(row=0, column=1, padx=5, pady=5)
end_combo.grid(row=1, column=1, padx=5, pady=5)

# Sonuç kutusu
output = tk.Text(root, height=4, width=100)
output.pack(pady=5)

# Grafik alanı
fig, ax = plt.subplots(figsize=(15, 10))
canvas = FigureCanvasTkAgg(fig, master=root)
canvas.get_tk_widget().pack(fill=tk.BOTH, expand=True)

# Harita çizimi
def harita_ciz(graf, yol=None):
    ax.clear()
    pos = sabit_konumlar_sutun(hatlar)
    nx.draw(graf, pos, with_labels=True, node_color='lightblue',
            edge_color='gray', node_size=1000, font_size=9, ax=ax)

    if yol:
        path_edges = list(zip(yol[:-1], yol[1:]))
        nx.draw_networkx_nodes(graf, pos, nodelist=yol, node_color='orange', ax=ax)
        nx.draw_networkx_edges(graf, pos, edgelist=path_edges, edge_color='red', width=3, ax=ax)

    canvas.draw()

# Buton fonksiyonu
def rota_goster():
    kaynak = start_var.get()
    hedef = end_var.get()
    output.delete(1.0, tk.END)

    if kaynak not in G.nodes or hedef not in G.nodes:
        output.insert(tk.END, "❌ Lütfen geçerli iki durak seçin.")
        return

    yol, uzunluk = en_kisa_yol(G, kaynak, hedef)
    if yol:
        sure = uzunluk * 2  # Ortalama 2 dakika/durak
        output.insert(tk.END, f"{kaynak} → {hedef} arası en kısa yol ({uzunluk} durak / yaklaşık {sure} dk):\n")
        output.insert(tk.END, " → ".join(yol))
        harita_ciz(G, yol)
    else:
        output.insert(tk.END, f"🚫 {kaynak} → {hedef} arasında yol bulunamadı.")

# Buton
ttk.Button(frame, text="Rota Göster", command=rota_goster).grid(row=2, column=0, columnspan=2, pady=10)

# İlk çizim
harita_ciz(G)

root.mainloop()
```
