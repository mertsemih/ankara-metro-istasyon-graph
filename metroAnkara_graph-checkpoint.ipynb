```python
import networkx as nx
import matplotlib.pyplot as plt

# Hatlardaki duraklar
A1 = ["AT", "Emek", "Bahelievler", "Beevler", "Maltepe", "Demirtepe", "Kzlay", "Kolej", "Kurtulu", "Cebeci", "Dikimevi"]
M1 = ["Kzlay", "Shhiye", "Adliye", "Ulus", "Atatrk Kltr Merkezi", "Akkpr", "vedik", "Yenimahalle", "Demetevler", "Hastane", "Macunky", "OSTM", "Batkent"]
M2 = ["Kzlay", "Necatibey", "Milli Ktphane", "Stz", "MTA", "Emek", "Bilkent", "Beytepe", "mitky", "ayyolu", "Koru"]
M3 = ["Batkent", "Bat Merkez", "Mesa", "Botanik", "stanbul Yolu", "Eryaman 5", "Eryaman 1-2", "Devlet Mah.", "Harikalar Diyar", "GOP", "OSB-Trekent"]
M4 = ["AKM", "ASK", "Dkap", "Meteoroloji", "Belediye", "Mecidiye", "Kuyuba", "Dutluk", "ehitler"]
BAKENTRAY = ["Sincan", "Lale", "Elvankent", "Eryaman YHT", "zgne", "Etimesgut", "Havadura", "Yldrm", "Behibey", "Motor", "Gazi", "Gazi Mahallesi", "Hipodrom", "Ankara Gar", "Kurtulu", "Cebeci", "Demirlibahe", "Saimekadn", "Mamak", "Baderesi", "reil", "Kstence", "Kayas"]
T1 = ["Yenimahalle", "vedik", "Yunus Emre", "TRT Seyir", "entepe"]

# Graf olutur
G = nx.Graph()

# Hat balantlarn ekleyen fonksiyon
def add_line(line):
    for i in range(len(line) - 1):
        G.add_edge(line[i], line[i + 1])

# Hatlar graf'a ekle
for line in [A1, M1, M2, M3, M4, BAKENTRAY, T1]:
    add_line(line)

# Grselletirme
plt.figure(figsize=(18, 12))
nx.draw(G, with_labels=True, node_color="skyblue", node_size=800, font_size=8)
plt.title("Ankara Metro Sistemi - Graph Teorisi ile Modelleme", fontsize=14)
plt.show()
```

```python
import numpy as np

# Dmleri liste halinde al
nodes = list(G.nodes())
nodes.sort()  # Alfabetik sralama (istee bal)

# Dm isimlerini indeksle
node_index = {node: i for i, node in enumerate(nodes)}

# Komuluk matrisi olutur
adj_matrix = np.zeros((len(nodes), len(nodes)), dtype=int)

# Kenarlar matrise ile
for u, v in G.edges():
    i, j = node_index[u], node_index[v]
    adj_matrix[i][j] = 1
    adj_matrix[j][i] = 1  # Simetrik nk ynsz grafik

# Matrisi yazdr
print("Komuluk Matrisi:")
print("    " + "  ".join([n[:4] for n in nodes]))  # lk 4 harfi gster
for i, row in enumerate(adj_matrix):
    print(f"{nodes[i][:4]} {row}")
```

```python
import networkx as nx

# En ksa yol fonksiyonu
def en_kisa_yol(graf, kaynak, hedef):
    try:
        yol = nx.shortest_path(graf, source=kaynak, target=hedef)
        uzunluk = nx.shortest_path_length(graf, source=kaynak, target=hedef)
        return yol, uzunluk
    except nx.NetworkXNoPath:
        return None, float('inf')

# rnek sorgu
kaynak_durak = "Kzlay"
hedef_durak = "Sincan"
yol, uzunluk = en_kisa_yol(G, kaynak_durak, hedef_durak)

# Sonular gster
print(f"{kaynak_durak}  {hedef_durak} aras en ksa yol ({uzunluk} durak):")
print("  ".join(yol))
```

```python
import tkinter as tk
from tkinter import ttk
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

# Metro hatlar
hatlar = {
    "A1": ["AT", "Emek", "Bahelievler", "Beevler", "Maltepe", "Demirtepe", "Kzlay", "Kolej", "Kurtulu", "Cebeci", "Dikimevi"],
    "M1": ["Kzlay", "Shhiye", "Adliye", "Ulus", "Atatrk Kltr Merkezi", "Akkpr", "vedik", "Yenimahalle", "Demetevler", "Hastane", "Macunky", "OSTM", "Batkent"],
    "M2": ["Kzlay", "Necatibey", "Milli Ktphane", "Stz", "MTA", "Emek", "Bilkent", "Beytepe", "mitky", "ayyolu", "Koru"],
    "M3": ["Batkent", "Bat Merkez", "Mesa", "Botanik", "stanbul Yolu", "Eryaman 5", "Eryaman 1-2", "Devlet Mah.", "Harikalar Diyar", "GOP", "OSB-Trekent"],
    "M4": ["Atatrk Kltr Merkezi", "ASK", "Dkap", "Meteoroloji", "Belediye", "Mecidiye", "Kuyuba", "Dutluk", "ehitler"],
    "BAKENTRAY": ["Sincan", "Lale", "Elvankent", "Eryaman YHT", "zgne", "Etimesgut", "Havadura", "Yldrm", "Behibey", "Motor", "Gazi", "Gazi Mahallesi", "Hipodrom", "Ankara Gar", "Kurtulu", "Cebeci", "Demirlibahe", "Saimekadn", "Mamak", "Baderesi", "reil", "Kstence", "Kayas"],
    "T1": ["Yenimahalle", "vedik", "Yunus Emre", "TRT Seyir", "entepe"]
}

# Graf yaps olutur
G = nx.Graph()
for hat in hatlar.values():
    for i in range(len(hat) - 1):
        G.add_edge(hat[i], hat[i + 1])

# Sabit konumlar stun gibi
def sabit_konumlar_sutun(hatlar_dict):
    pos = {}
    x_step = 5
    y_step = -1
    x = 0
    for hat in hatlar_dict.values():
        for i, durak in enumerate(hat):
            if durak not in pos:
                pos[durak] = (x, i * y_step)
        x += x_step
    return pos

# En ksa yolu bul
def en_kisa_yol(graf, kaynak, hedef):
    try:
        yol = nx.shortest_path(graf, source=kaynak, target=hedef)
        return yol, len(yol) - 1
    except nx.NetworkXNoPath:
        return None, None

# Arayz
root = tk.Tk()
root.title("Ankara Metro - Rota ve Harita")
root.state('zoomed')  # Tam ekran

frame = ttk.Frame(root)
frame.pack(pady=20)

ttk.Label(frame, text="Balang Dura:").grid(row=0, column=0, padx=5, pady=5)
ttk.Label(frame, text="Var Dura:").grid(row=1, column=0, padx=5, pady=5)

duraklar = sorted(G.nodes())
start_var = tk.StringVar()
end_var = tk.StringVar()
start_combo = ttk.Combobox(frame, textvariable=start_var, values=duraklar, width=30)
end_combo = ttk.Combobox(frame, textvariable=end_var, values=duraklar, width=30)
start_combo.grid(row=0, column=1, padx=5, pady=5)
end_combo.grid(row=1, column=1, padx=5, pady=5)

# Sonu kutusu
output = tk.Text(root, height=4, width=100)
output.pack(pady=5)

# Grafik alan
fig, ax = plt.subplots(figsize=(15, 10))
canvas = FigureCanvasTkAgg(fig, master=root)
canvas.get_tk_widget().pack(fill=tk.BOTH, expand=True)

# Harita izimi
def harita_ciz(graf, yol=None):
    ax.clear()
    pos = sabit_konumlar_sutun(hatlar)
    nx.draw(graf, pos, with_labels=True, node_color='lightblue',
            edge_color='gray', node_size=1000, font_size=9, ax=ax)

    if yol:
        path_edges = list(zip(yol[:-1], yol[1:]))
        nx.draw_networkx_nodes(graf, pos, nodelist=yol, node_color='orange', ax=ax)
        nx.draw_networkx_edges(graf, pos, edgelist=path_edges, edge_color='red', width=3, ax=ax)

    canvas.draw()

# Buton fonksiyonu
def rota_goster():
    kaynak = start_var.get()
    hedef = end_var.get()
    output.delete(1.0, tk.END)

    if kaynak not in G.nodes or hedef not in G.nodes:
        output.insert(tk.END, " Ltfen geerli iki durak sein.")
        return

    yol, uzunluk = en_kisa_yol(G, kaynak, hedef)
    if yol:
        sure = uzunluk * 2  # Ortalama 2 dakika/durak
        output.insert(tk.END, f"{kaynak}  {hedef} aras en ksa yol ({uzunluk} durak / yaklak {sure} dk):\n")
        output.insert(tk.END, "  ".join(yol))
        harita_ciz(G, yol)
    else:
        output.insert(tk.END, f" {kaynak}  {hedef} arasnda yol bulunamad.")

# Buton
ttk.Button(frame, text="Rota Gster", command=rota_goster).grid(row=2, column=0, columnspan=2, pady=10)

# lk izim
harita_ciz(G)

root.mainloop()
```
